```python
from UROP_matching import get_hs_code
import pandas as pd
import tqdm
```

```python
# open the 1789HS.xlsx file
df = pd.read_excel('1789HS.xlsx')
# get all brief descriptions into a list
brief_descriptions = df['brief_description'].tolist()

len(brief_descriptions)
```

```python
brief_descriptions[22]

# remove brief_descriptions[22] from brief_descriptions
brief_descriptions.remove(brief_descriptions[22])
len(brief_descriptions)
```

```python
from tqdm import tqdm

# create a map between brief descriptions and HS codes using the get_hs_code function
brief_description_to_hs_code = {}
hs_code_to_product = {}
for brief_description in tqdm(brief_descriptions, desc="Processing brief descriptions"):
    # run the get_hs_code function at most 3 times if the function fails
    for i in range(3):
        try:
            hs_code, product = get_hs_code(brief_description)
            break
        except:
            continue
    brief_description_to_hs_code[brief_description] = hs_code
    hs_code_to_product[hs_code] = product
```

```python
# export brief_description_to_hs_code to a pickle file
import pickle

with open('brief_description_to_hs_code.pickle', 'wb') as handle:
    pickle.dump(brief_description_to_hs_code, handle, protocol=pickle.HIGHEST_PROTOCOL)

# export hs_code_to_product to a pickle file
with open('hs_code_to_product.pickle', 'wb') as handle:
    pickle.dump(hs_code_to_product, handle, protocol=pickle.HIGHEST_PROTOCOL)

    
```

```python
# brief_description_to_hs_code has the brief_description as the key and the hs_code as the value
# need to add a column to df called new_hs and fill it with the hs_code values given the brief_description from the brief_description_to_hs_code dictionary
df['new_hs'] = df['brief_description'].map(brief_description_to_hs_code)

# make a new column called new_product and fill it with the product values given the hs_code from the hs_code_to_product dictionary
df['new_product'] = df['new_hs'].map(hs_code_to_product)

# only keep columns year, hs, brief_description, new_hs, new_product from df
new_df = df[['year', 'hs', 'brief_description', 'new_hs', 'new_product']]
```

```python
# need to fix the new_product column. each entry in that column follows format [hs_code: product]. need to just extract the product
# remove new_product2 from new_df
new_df = new_df.drop(columns=['new_product2'])
```

```python
# export new_df as a pickle file
import pickle
with open('new_df.pkl', 'wb') as f:
    pickle.dump(new_df, f)

```

```python
new_df['new_product2'] = new_df['new_product'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
```

```python
# each entry in new_product2 is represented as a string where it is 'hs_code: product'. We just want the product after the :
# add the product in a new column

new_df['new_product2'] = new_df['new_product2'].apply(lambda x: x.split(': ')[1] if x is not None else None)
```

```python
# run f1 score for 2 digit between column hs and new_hs
# the value for a given row should be 1 if the first 2 digits of hs and new_hs are the same, 0 otherwise
# add a column to the new_df called f1

# first, we need to make sure the numbers in hs are two digits. if not, we need to add a 0 in front of the number
new_df['hs'] = new_df['hs'].apply(lambda x: '0' + str(x) if len(str(x)) == 1 else str(x))

new_df['f1'] = new_df.apply(lambda row: 1 if str(row['hs'])[:2] == str(row['new_hs'])[:2] else 0, axis=1)
```

```python
f1_score = new_df['f1'].sum() / (len(new_df) -1)

f1_score
```
